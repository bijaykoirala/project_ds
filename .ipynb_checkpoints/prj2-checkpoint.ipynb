```python
# Getting the data from the folder and making a dict

data_url = "http://apps.start.umd.edu/gtd/downloads/dataset/globalterrorismdb_0615dist.xlsx"
data_file = "data/globalterrorismdb_0615dist.xlsx"
data_file_in_csv = "data/globalterrorismdb_0615dist.csv"

# if the file does not exist, download that
if not os.path.exists(data_file):
    print "--- Downloading files ---"
    urllib.urlretrieve(data_url, filename=data_file)
    print "--- Download Complete ---"
else:
    print("data already downloaded")
```

After getting the dataset, we found out that the dataset was an xlsx format file, so there were two options, either to save it as a csv using microsoft excel or try to deal with the same file format. Let's try the later first. I downloaded <u>xlrd</u> package from the package manager. After a few minutes, we were able to produce a csv file which we will use throughout the project.

```python
if not os.path.exists(data_file_in_csv):
    filepath = 'data'
    wb = xlrd.open_workbook(os.path.join(filepath, 'globalterrorismdb_0615dist.xlsx'))
    sheet = wb.sheet_by_index(0)
    fp = open(os.path.join(filepath, 'globalterrorismdb_0615dist.csv'), 'wb')
    wr = csv.writer(fp, quoting=csv.QUOTE_ALL)
    for rownum in xrange(sheet.nrows):
          wr.writerow([unicode(val).encode('utf8') for val in sheet.row_values(rownum)])
    fp.close()
else:
    print("Data already set")
```

Now that the data is all with us, let us read the data nicely in a list which we can reuse throughout the project.

```python
import csv
f = open(data_file_in_csv, 'r')
csvDict = csv.DictReader(f) 
all_attacks = []
for row in csvDict:
    all_attacks.append(row)
f.close()

import pprint
print(len(all_attacks), "attacks in the database")
print(len(all_attacks[6].keys()), "headers in the database")
```

# What is the trend of terrorism in the world right now?

Is terrorism in the rise or in the fall? We want to see throughout different regions of the world.

```python
list_attacks_years = []
for each_attack in all_attacks:
    list_attacks_years.append(float(each_attack["iyear"]))

count_years = Counter(list_attacks_years)

list_years = count_years.keys()
list_attacks = count_years.values()
plt.plot(list_years, list_attacks , 'o-', c=niceblu)
plt.xlabel("Years"); 
plt.ylabel("Number of terrorist activities");
plt.title("Trend of global terrorist activities");
plt.gcf().set_size_inches(9, 6)
plt.savefig("int/global_trend_terrorism.png")
plt.show()
```

Now lets see if the terrorists have a favorite season for their activities

Terrorist activities started off at a mild pace, they were gradually increasing and even declined during 2000s but started quickly rising since 2005. They are sky rocketing at the present situation.


```python
list_attacks_months = []
for each_attack in all_attacks:
    if (float(each_attack["imonth"]) > 0):
        list_attacks_months.append(float(each_attack["imonth"]))

count_months = Counter(list_attacks_months)

list_months = count_months.keys()
print(list_months)
list_attacks = count_months.values()
plt.bar(list_months, list_attacks , align='center', facecolor = niceblu)
plt.xlabel("month"); 
plt.ylabel("Number of terrorist activities");
plt.xticks(list_months, [ "Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug","Sep","Oct","Nov","Dec"])
plt.title("Trend of global terrorism activities");
plt.gcf().set_size_inches(9, 6)
plt.savefig("int/global_trend_terrorism_month.png")
plt.show()
```

This distribution does not look like it is affected by any trend

```python
def get_attack_year_for_region(region):
    """returns tuples of list of years and list of attacks during the period"""
    list_attacks_years = []
    for each_attack in all_attacks:
        if (each_attack["region_txt"] == region):
            list_attacks_years.append(float(each_attack["iyear"]))

    count_years = Counter(list_attacks_years)

    list_years = count_years.keys()
    list_attacks = count_years.values()
    return list_years, list_attacks
```

Now let us first print each of these and plot them to see the trends.

```python
# get the list of all regions
set_all_regions = set()
for each_attack in all_attacks:
    set_all_regions.add(each_attack["region_txt"])
    

for each_region in set_all_regions:
    list_years, list_attacks = get_attack_year_for_region(each_region)
    plt.plot(list_years, list_attacks , 'o-', c=niceblu)
    plt.xlabel("Years"); 
    plt.ylabel("Number of terrorist activities");
    plt.title("Trend of terrorism activities in "+each_region);
    plt.gcf().set_size_inches(9, 6)
    plt.savefig("int/"+each_region+" terrorism.png")
    plt.show()
```

```python
def get_attack_year_for_country_after_2k(country):
    """returns tuples of list of years and list of attacks during the period"""
    list_attacks_years = []
    for each_attack in all_attacks:
        if (each_attack["country_txt"] == country):
            year = (float(each_attack["iyear"]))
            if year > 1999:
                list_attacks_years.append(float(each_attack["iyear"]))

    count_years = Counter(list_attacks_years)

    list_years = count_years.keys()
    list_attacks = count_years.values()
    return list_years, list_attacks
```

# Have the terrorist attacks been less succssful over the years?

To answer this question, we need to look at not unsucessful attacks but also attacks that were successful. We calculate percentages of attacks we could prevent over the course of years and plot them to have an understanding of what we achieved.

```python
successful_yrs, failed_years = [], []
for each_attack in all_attacks:
    if float(each_attack["success"])==1.0:
        successful_yrs.append(float(each_attack["iyear"]))
    elif float(each_attack["success"])==0.0:
        failed_years.append(float(each_attack["iyear"]))
    else:
        print(each_attack["success"])


count_syears = Counter(successful_yrs)
count_fyears = Counter(failed_years)

list_years = count_years.keys()
list_svalues = count_syears.values()
list_fvalues = count_fyears.values();
list_percent = []

for index in range(len(list_svalues)):
    success = list_svalues[index]
    failure = list_fvalues[index]
    list_percent.append(failure*100/(success+failure))
    

print(list_years)

plt.plot(list_years, list_percent , 'o-',c=niceblu)
plt.xlabel("Years"); 
plt.ylabel("% of Successful Preventions of Attacks");
plt.title("Global trend of % of successfully prevented attacks");
plt.gcf().set_size_inches(9, 6)
plt.gca().set_ylim([min(list_percent)-1, max(list_percent)+1])
plt.savefig("int/global_trend_successful_prevents.png")
plt.show()
```

Conclusion: Looks like we have never been very successful to prevent terrorist attacks since the early days. The success rate looks to be fluctuating all these years from 70 to 2005. In the recent yeras, however, after around 2007, the success rate for preventing terrorist attacks to have fallen to significantly improved to the past. A better military might be accountable for this. The final answer is we look like we slightly have been able to prevent attacks. 

Now, lets look at where are the places we have been able to prevent attacks mostly on.

```python
successful_loc, failed_loc = [], []
for each_attack in all_attacks:
    lat, lon = each_attack["latitude"], each_attack["longitude"]
    if len(lat)>0 and len(lon)>0 and float(each_attack["success"])==1.0:
        trim_lat = lat[0:lat.find(".")+3]
        trim_lon = lon[0:lon.find(".")+3]
        successful_loc.append((trim_lat, trim_lon))
    elif len(lat)>0 and len(lon)>0 and float(each_attack["success"])==0.0:
        trim_lat = lat[0:lat.find(".")+3]
        trim_lon = lon[0:lon.find(".")+3]
        failed_loc.append((trim_lat, trim_lon))


count_s = Counter(successful_loc)
count_f = Counter(failed_loc)
list_success_percent = []

for k,v in count_s.iteritems():
    if k in count_f:
        v_fail = count_f[k]
    else:
        v_fail = 0
    percent = v_fail*100/(v+v_fail)
    list_success_percent.append((k,percent))

refined_list = []
for each_item in list_success_percent:
    if each_item[1]>1:
        refined_list.append(each_item)
print(len(refined_list))
```

```python
def get_suicide_attack_year_for_region(region):
    """returns tuples of list of years and list of attacks during the period"""
    list_attacks_years = []
    for each_attack in all_attacks:
        if (each_attack["region_txt"] == region) and each_attack["suicide"]=="1.0":
            list_attacks_years.append(float(each_attack["iyear"]))

    count_years = Counter(list_attacks_years)

    list_years = count_years.keys()
    list_attacks = count_years.values()
    return list_years, list_attacks
```

```python
news_words = []
for each_attack in all_attacks:
    motive = each_attack["motive"]
    filter_list = ["the", "is", "for", "was", "and", "by", "were", "on", "as", "to", "with", "may", 
                   "who", "of", "that", "at", "in", "a", "is", "be"]
    if motive != "":
        for ew in motive.split(' '):
            ew = ew.lower()
            if "." in ew:
                ew = ew.replace(".", '')
            if ";" in ew:
                ew = ew.replace(";", "")
            #removing articles and words like prepositions
            if ew not in filter_list:
                news_words.append(ew)
fOut = open("wordcloud.txt", 'w')
sth = 1
for w in Counter(news_words).most_common(100):
    for stuff in range(sth):
        fOut.write("\""+str(w[0])+"\", ")
    #sth = sth - 1
fOut.close()
print(len(news_words))
# we will create a word cloud using either d3 or some python library should time permit.
# But for now, moving on to join the data

                
```

```python
from numpy import interp 
scaled_range = [0,100] # the range for final comparision 

list_attacks_years = []
for each_attack in all_attacks:
    list_attacks_years.append(float(each_attack["iyear"]))

# we have to scale list_attacks so that all fields of comparisions have similar grounds
count_years = Counter(list_attacks_years)
scaled_attacks = []

list_years = count_years.keys()
list_attacks = count_years.values()
for each in list_attacks:
    scaled_attacks.append(interp(each,[min(count_years.values()),max(count_years.values())], scaled_range))

plt.plot(list_years, scaled_attacks , 'o-')

# for multiple attacks,
years = []
for each_attack in all_attacks:
    if float(each_attack["multiple"])==1.0:
        years.append(float(each_attack["iyear"]))

count_years = Counter(years)
scaled_attacks = []
list_years = count_years.keys()
list_attacks = count_years.values()
list_years = count_years.keys()
list_attacks = count_years.values()
for each in list_attacks:
    scaled_attacks.append(interp(each,[min(count_years.values()),max(count_years.values())], scaled_range))

# multiple attacks removed to reduce crowding
#plt.plot(list_years, scaled_attacks , 'o-')

# for successful attacks
successful_yrs, failed_years = [], []
for each_attack in all_attacks:
    if float(each_attack["success"])==1.0:
        successful_yrs.append(float(each_attack["iyear"]))
    elif float(each_attack["success"])==0.0:
        failed_years.append(float(each_attack["iyear"]))
    else:
        print(each_attack["success"])


count_syears = Counter(successful_yrs)
count_fyears = Counter(failed_years)

list_years = count_years.keys()
list_svalues = count_syears.values()
list_fvalues = count_fyears.values();
list_percent = []

for index in range(len(list_svalues)):
    success = list_svalues[index]
    failure = list_fvalues[index]
    list_percent.append(success*100/(success+failure))
    
plt.plot(list_years, list_percent , 'o-')

# for suicide attacks
list_attacks_years = []
for each_attack in all_attacks:
    if each_attack["suicide"]=="1.0":
        list_attacks_years.append(float(each_attack["iyear"]))

count_years = Counter(list_attacks_years)

list_years = count_years.keys()
list_attacks = count_years.values()
scaled_attacks = []
for each in list_attacks:
    scaled_attacks.append(interp(each,[min(count_years.values()),max(count_years.values())], scaled_range))

plt.plot(list_years, scaled_attacks , 'o-')

# deaths due to terrorist activities
dict_year_deaths = {}
for each_attack in all_attacks:
    year = float(each_attack["iyear"])
    try:
        deaths = float(each_attack["nkill"])
    except ValueError:
        continue
    else:

        if year in dict_year_deaths:
            dict_year_deaths[year] += deaths
        else:
            dict_year_deaths[year] = deaths


scaled_attacks = []
for each in dict_year_deaths.values():
    scaled_attacks.append(interp(each,[min(dict_year_deaths.values()),max(dict_year_deaths.values())], scaled_range))

plt.plot(dict_year_deaths.keys(), scaled_attacks , 'o-')
#plt.plot(dict_year_deaths.keys(), dict_year_deaths.values() , 'o-', c = nicegrn)

plt.xlabel("Years"); 
plt.ylabel("Scaled values (0-100)");
plt.ylim(0,110)
plt.title("Global trend of terrorist activities");
plt.gcf().set_size_inches(9, 6)

plt.legend(["No. of Attacks","Success Rate", "Suicide Attacks", "No. of Deaths"], loc="middle left")
plt.savefig("int/trend_global.png")
plt.show()
```
