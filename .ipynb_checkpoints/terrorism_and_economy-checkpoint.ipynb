```python
# Getting the data from the folder and making a dict

data_url = "http://apps.start.umd.edu/gtd/downloads/dataset/globalterrorismdb_0615dist.xlsx"
data_file = "data/globalterrorismdb_0615dist.xlsx"
data_file_in_csv = "data/globalterrorismdb_0615dist.csv"

# if the file does not exist, download that
if not os.path.exists(data_file):
    print "--- Downloading files ---"
    urllib.urlretrieve(data_url, filename=data_file)
    print "--- Download Complete ---"
else:
    print("data already downloaded")
```

After getting the dataset, we found out that the dataset was an xlsx format file, so there were two options, either to save it as a csv using microsoft excel or try to deal with the same file format. Let's try the later first. I downloaded <u>xlrd</u> package from the package manager. After a few minutes, we were able to produce a csv file which we will use throughout the project.

```python
if not os.path.exists(data_file_in_csv):
    filepath = 'data'
    wb = xlrd.open_workbook(os.path.join(filepath, 'globalterrorismdb_0615dist.xlsx'))
    sheet = wb.sheet_by_index(0)
    fp = open(os.path.join(filepath, 'globalterrorismdb_0615dist.csv'), 'wb')
    wr = csv.writer(fp, quoting=csv.QUOTE_ALL)
    for rownum in xrange(sheet.nrows):
          wr.writerow([unicode(val).encode('utf8') for val in sheet.row_values(rownum)])
    fp.close()
else:
    print("Data already set")
```

Now that the data is all with us, let us read the data nicely in a list which we can reuse throughout the project.

```python
import csv
f = open(data_file_in_csv, 'r')
csvDict = csv.DictReader(f) 
all_attacks = []
for row in csvDict:
    all_attacks.append(row)
f.close()

import pprint
print(len(all_attacks), "attacks in the database")
print(len(all_attacks[6].keys()), "headers in the database")
```

```python
countries_terr = {}
for each_attack in all_attacks:
    country = each_attack["country_txt"]
    if country not in countries_terr:
        countries_terr[country] = {}
        
    else:
        year = float(each_attack["iyear"])
        current_country = countries_terr[country]
        if year not in current_country:
            current_country[year] = 1
        else:
            current_country[year] += 1

print(countries_terr.get("Iraq", "Data not found"))
```

Now, lets do the same for our health dataset. For this one, we do not have the links because the process was manual. The process was explained on the end of prj2.ipynb

```python
import csv
f = open("data/health_wb.csv", 'r')
csvDict = csv.DictReader(f) 
all_health = []
for row in csvDict:
    all_health.append(row)
f.close()

print(len(all_health), "records in the database")
print(len(all_health[6].keys()), "headers in the database")
```

Lets do a plot for health  expenditure for one country to see the data

```python
def generate_dataset_for(series_name):
    d_health_expenditures = {}
    for item in all_health:
        if item["Series Name"]==series_name:
            country = item["\xef\xbb\xbfCountry Name"]
            if country not in d_health_expenditures:
                d_health_expenditures[country] = {}
                for each_item in item:
                    #print(each_item)
                    if "[YR" in each_item:
                        year = each_item[0:4]
                        year = float(year)
                        value = item[each_item]
                        if ".." in value:
                            value = 0.0
                            continue
                        else:
                            value = float(value)
                        d_health_expenditures[country][year] = value
    return d_health_expenditures
```

The data structure was a little more complicated and required filtering for country name, year was also not in a proper format. So, we has to do a little filtering and looping to get the things in the way. We modified the function so that it can accomodate any series we want.

```python
dict_health_expenditures = generate_dataset_for("Health expenditure, total (% of GDP)")
```

```python
health_nepal =  dict_health_expenditures['Nepal']
terror_nepal =  countries_terr['Nepal']
normalized_nepal_health = [float(x)/max(health_nepal.values()) for x in health_nepal.values()]
normalized_nepal_terror = [float(x)/max(terror_nepal.values()) for x in terror_nepal.values()]
plt.plot(health_nepal.keys(), normalized_nepal_health , 'o-')
plt.plot(terror_nepal.keys(), normalized_nepal_terror, 'o-')
plt.xlabel("Years"); 
plt.ylabel("Health expenditure");
plt.title("");
plt.gcf().set_size_inches(9, 6)
plt.savefig("int/health_exp_nepal.png")
plt.title("Comparision of normalied health and terrorist activities in Nepal")
plt.legend(["health", "terrorist activities"], loc="upper left")
plt.show()
```

# #Analyzing p- value and r squared remains ...

Lets replicate these for all the countries and lets collect the lists of R values and p-values

```python
# First we should do something to make sure that the countries we are looking for are in both sets
s_h_countries = set()
s_t_countries = set()
for item in all_health:
    country = item["\xef\xbb\xbfCountry Name"]
    s_h_countries.add(country)
    if '' in s_h_countries:
        s_h_countries.remove('')
for item in all_attacks:
    country = item["country_txt"]
    s_t_countries.add(country)
    
s_both_countries = s_h_countries.intersection(s_t_countries)
```

```python
dict_r_p_values = {}
for country in s_both_countries:
    health_data = dict_health_expenditures[country]
    terror_data = countries_terr[country]
    
    # we should now make sure the years are also same for both of these
    sHealth = set(health_data.keys())
    sTerror = set(terror_data.keys())
    sBoth = sHealth.intersection(sTerror)
    l_health, l_terror = [], []
    for year in sBoth:
        l_health.append(health_data[year])
        l_terror.append(terror_data[year])
    norm_health = [float(x)/max(l_health) for x in l_health]
    norm_terror = [float(x)/max(l_terror) for x in l_terror]

    # now we want to see if there is some kind of relationship between the two of these
    if len(norm_health)> 1:
        beta1, beta0, r, p, se = scipy.stats.linregress(l_health, l_terror)
        #print " Coeffs =", beta0,beta1
        if math.isnan(r) or math.isnan(p):
            continue
        dict_r_p_values[country] = {"r-sq":r**2, "p-val":p}
        
        
# r-values, p_values and country
rvalues, pvalues, countries = [],[],[] 
total_rval , total_pval = 0,0
for k,v in dict_r_p_values.iteritems():
    r_sq = v["r-sq"]
    p_va = v["p-val"]
    total_rval += r_sq
    total_pval += p_va
    rvalues.append(r_sq)
    pvalues.append(p_va)
    countries.append(k)
    #print(format(k, '20'), format(r_sq, '10.7f'), format(p_va, '10.7f'))
    
print("Average r-value : ",total_rval/len(dict_r_p_values.keys()))
print("Average p-value : ",total_pval/len(dict_r_p_values.values()))


print(sort(pvalues))
    
```

The average of the two values could not provide us with any conclusive answer. A better idea would be to just consider the countries that have highest number of terrorist attacks



Lets see the patterns for the top 5 countries that observe the highest terrorist activities:-

```python
def get_attack_year_for_country_after_2k(country):
    """returns tuples of list of years and list of attacks during the period"""
    list_attacks_years = []
    for each_attack in all_attacks:
        if (each_attack["country_txt"] == country):
            year = (float(each_attack["iyear"]))
            if year > 1999:
                list_attacks_years.append(float(each_attack["iyear"]))

    count_years = Counter(list_attacks_years)

    list_years = count_years.keys()
    list_attacks = count_years.values()
    return list_years, list_attacks

attacks_yrs = []
for each_attack in all_attacks:
    year = (float(each_attack["iyear"]))
    if year>2004:
        attacks_yrs.append(each_attack["country_txt"])
print(len(attacks_yrs))

# top 5
list_yrs_iraq, list_incidents_iraq = get_attack_year_for_country_after_2k("Iraq")
list_yrs_paki, list_incidents_paki = get_attack_year_for_country_after_2k("Pakistan")
list_yrs_afgani, list_incidents_afgani = get_attack_year_for_country_after_2k("Afghanistan")
list_yrs_india, list_incidents_india = get_attack_year_for_country_after_2k("India")
list_yrs_thailand, list_incidents_thailand = get_attack_year_for_country_after_2k("Thailand")


# plotting top 5
plt.plot(list_yrs_iraq, list_incidents_iraq , 'o-')
plt.plot(list_yrs_paki, list_incidents_paki , 'o-')
plt.plot(list_yrs_afgani, list_incidents_afgani , 'o-')
plt.plot(list_yrs_india, list_incidents_india , 'o-')
plt.plot(list_yrs_thailand, list_incidents_thailand , 'o-')

plt.xlabel("Years"); 
plt.ylabel("Number of terrorist activities in the past decade");
plt.title("Trend of terrorism on most influenced countries");
plt.gcf().set_size_inches(8,6)
plt.savefig("int/top_5_terrorist_attacks_countries_past_decade.png")

plt.legend(["Iraq", "Pakistan", "Afganistan", "India","Thailand"], loc="upper left", fontsize=12)
plt.show()
```

```python
#a function to return r-values
def get_r_p(health_data, teror_data):
    sHealth = set(health_data.keys())
    sTerror = set(terror_data.keys())
    sBoth = sHealth.intersection(sTerror)
    l_health, l_terror = [], []
    for year in sBoth:
        l_health.append(health_data[year])
        l_terror.append(terror_data[year])
    norm_health = [float(x)/max(l_health) for x in l_health]
    norm_terror = [float(x)/max(l_terror) for x in l_terror]

    # now we want to see if there is some kind of relationship between the two of these
    if len(norm_health)> 1:
        beta1, beta0, r, p, se = scipy.stats.linregress(l_health, l_terror)
        return r**2
```

# Cause of death by Injury analysis

```python
dict_rd = generate_dataset_for("Cause of death, by injury (% of total)")
#print(dict_rd)
#a function to return r-values
def get_r_p(injury_data, teror_data):
    sinjury = set(injury_data.keys())
    sTerror = set(terror_data.keys())
    sBoth = sinjury.intersection(sTerror)
    l_injury, l_terror = [], []
    for year in sBoth:
        l_injury.append(injury_data[year])
        l_terror.append(terror_data[year])
    norm_injury = [float(x)/max(l_injury) for x in l_injury]
    norm_terror = [float(x)/max(l_terror) for x in l_terror]

    # now we want to see if there is some kind of relationship between the two of these
    if len(norm_injury)> 1:
        beta1, beta0, r, p, se = scipy.stats.linregress(l_injury, l_terror)
        return r**2
    
fig = plt.figure(figsize=(12,12))
markerSize =5
dict_health = dict_rd

#For IRAQ
ax = plt.subplot(321)
data_country =  dict_health['Iraq']
terror_country =  countries_terr['Iraq']
plt.plot(data_country.keys(), data_country.values(), '.-', c=niceblu)
fig.subplots_adjust(hspace=0.25)
plt.legend(["Iraq"], loc="bottom right")
plt.title("Injury rates");

ax = plt.subplot(322)
plt.plot(terror_country.keys(), terror_country.values(), '.-', c=niceblu)
print("Iraq: ", get_r_p(data_country, terror_country))
plt.legend(["Iraq"], loc="bottom left")
plt.title("Terrorist Activities");
fig.subplots_adjust(hspace=0.25)

# repeat for Afganistan
ax = plt.subplot(323)
data_country =  dict_health['Afghanistan']
terror_country =  countries_terr['Afghanistan']
plt.plot(data_country.keys(), data_country.values(), '.-', c=niceblu)
plt.legend(["Afgan"], loc="top right")
fig.subplots_adjust(hspace=0.25)

ax = plt.subplot(324)
plt.plot(terror_country.keys(), terror_country.values(), '.-', c=niceblu)
print("Afganistan: ", get_r_p(data_country, terror_country))
plt.legend(["Afganistan"], loc="bottom left")
fig.subplots_adjust(hspace=0.5)

#For Pakistan
ax = plt.subplot(325)
data_country =  dict_health['Pakistan']
terror_country =  countries_terr['Pakistan']
plt.plot(data_country.keys(), data_country.values(), '.-', c=niceblu)
fig.subplots_adjust(hspace=0.25)
plt.legend(["Pakistan"], loc="bottom right")

ax = plt.subplot(326)
plt.plot(terror_country.keys(), terror_country.values(), '.-', c=niceblu)
print("Pakistan: ", get_r_p(data_country, terror_country))
plt.legend(["Pakistan"], loc="bottom left")
fig.subplots_adjust(hspace=0.25)

plt.show()
```
