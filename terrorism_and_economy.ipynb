After getting the dataset, we found out that the dataset was an xlsx format file, so there were two options, either to save it as a csv using microsoft excel or try to deal with the same file format. Let's try the later first. I downloaded <u>xlrd</u> package from the package manager. After a few minutes, we were able to produce a csv file which we will use throughout the project.

```python
if not os.path.exists(data_file_in_csv):
    filepath = 'data'
    wb = xlrd.open_workbook(os.path.join(filepath, 'globalterrorismdb_0615dist.xlsx'))
    sheet = wb.sheet_by_index(0)
    fp = open(os.path.join(filepath, 'globalterrorismdb_0615dist.csv'), 'wb')
    wr = csv.writer(fp, quoting=csv.QUOTE_ALL)
    for rownum in xrange(sheet.nrows):
          wr.writerow([unicode(val).encode('utf8') for val in sheet.row_values(rownum)])
    fp.close()
else:
    print("Data already set")
```

Now that the data is all with us, let us read the data nicely in a list which we can reuse throughout the project.

```python
import csv
f = open(data_file_in_csv, 'r')
csvDict = csv.DictReader(f) 
all_attacks = []
for row in csvDict:
    all_attacks.append(row)
f.close()

import pprint
print(len(all_attacks), "attacks in the database")
print(len(all_attacks[6].keys()), "headers in the database")
```

```python
countries_terr = {}
for each_attack in all_attacks:
    country = each_attack["country_txt"]
    if country not in countries_terr:
        countries_terr[country] = {}
        
    else:
        year = float(each_attack["iyear"])
        current_country = countries_terr[country]
        if year not in current_country:
            current_country[year] = 1
        else:
            current_country[year] += 1

print(countries_terr.get("Iraq", "Data not found"))
```

```python
def generate_dataset_for(series_name):
    d_economy = {}
    for item in all_economy:
        if item["Series Name"]==series_name:
            country = item["\xef\xbb\xbfCountry Name"]
            if country not in d_economy:
                d_economy[country] = {}
                for each_item in item:
                    #print(each_item)
                    if "[YR" in each_item:
                        year = each_item[0:4]
                        year = float(year)
                        value = item[each_item]
                        if ".." in value:
                            value = 0.0
                            continue
                        else:
                            value = float(value)
                        d_economy[country][year] = value
    return d_economy
```

Let us now see if more property is being damaged due to terrorist activities. To do that, let us assign weights to each value, if the attack is less than 1 mil, we can give the attack a value of 1, else if the attack costed more than a mil, we can give it a value of 10 and a value of 100 if the attack involved damages of overa a billion dollars.

```python
dict_damage = {}
for each in all_attacks:
    try:
        damage = float(each["propextent"])
        year = float(each["iyear"])
        if year in dict_damage:
            if damage == 3.0:
                dict_damage[year] += 1
            elif damage == 2.0:
                dict_damage[year] += 10
            elif damage == 1.0:
                dict_damage[year] += 1000
        else:
            if damage == 3.0:
                dict_damage[year] = 1
            elif damage == 2.0:
                dict_damage[year] = 10
            elif damage == 1.0:
                dict_damage[year] = 100
            
    except ValueError:
        continue
print(dict_damage)
```

A quick dirty plot to see if the trend is rising or falling !!!

```python
plt.plot(dict_damage.keys(), dict_damage.values() , 'o-',c=niceblu)
plt.xlabel("Years"); 
plt.ylabel("Property Damage in (in million dollars) ");
plt.title("Global trend of property damage");
plt.gcf().set_size_inches(9, 6)

plt.savefig("int/global_trend_property damage.png")
plt.show()
```

The chart shows the trend of property damage. Since we did not have exact damage accessment of all data, we categorized data into less than 1 million, between a million and a billion and more than a billion and awarded them values of 1 , 10 and 100 to standardize the expenses. The trend seems to be steadily growing since 2005. 2001 Saw a smaller peak (probably because of the 9/11 incident) but the current situation suggests something unusual. Between  2010 and 2015, the global damage due to terrorist activities has increased by at least 3 folds!
