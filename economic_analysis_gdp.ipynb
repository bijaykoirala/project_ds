After getting the dataset, we found out that the dataset was an xlsx format file, so there were two options, either to save it as a csv using microsoft excel or try to deal with the same file format. Let's try the later first. I downloaded <u>xlrd</u> package from the package manager. After a few minutes, we were able to produce a csv file which we will use throughout the project.

```python
if not os.path.exists(data_file_in_csv):
    filepath = 'data'
    wb = xlrd.open_workbook(os.path.join(filepath, 'globalterrorismdb_0615dist.xlsx'))
    sheet = wb.sheet_by_index(0)
    fp = open(os.path.join(filepath, 'globalterrorismdb_0615dist.csv'), 'wb')
    wr = csv.writer(fp, quoting=csv.QUOTE_ALL)
    for rownum in xrange(sheet.nrows):
          wr.writerow([unicode(val).encode('utf8') for val in sheet.row_values(rownum)])
    fp.close()
else:
    print("Data already set")
```

Now that the data is all with us, let us read the data nicely in a list which we can reuse throughout the project.

```python
def generate_dataset_for(series_name):
    d_economy = {}
    for item in all_economy:
        if item["Series Name"]==series_name:
            country = item["\xef\xbb\xbfCountry Name"]
            if country not in d_economy:
                d_economy[country] = {}
                for each_item in item:
                    #print(each_item)
                    if "[YR" in each_item:
                        year = each_item[0:4]
                        year = float(year)
                        value = item[each_item]
                        if ".." in value:
                            value = 0.0
                            continue
                        else:
                            value = float(value)
                        d_economy[country][year] = value
    return d_economy
```

The GDP of these countries seem to be either stagnant or falling. Alternately, now let us find out the countries with highest values of GDP and lets search if they have high terrorism activities or not.

```python
# find the highest ranking GDPs
gdp_2014 = {}
for each in all_economy:
    country = each["\xef\xbb\xbfCountry Name"]
    y_2014  = each["2014 [YR2014]"]
    y_2004  = each["2004 [YR2004]"]
    series = each["Series Name"]
    #print(series, country, y_2014)
    if series == "GDP per capita growth (annual %)" and y_2014 != ".." and y_2014 != "..":
        try:
            gdp_2014[country] = float(y_2014) - float(y_2004)
        except ValueError:
            continue
import operator
sorted_dict = sorted(gdp_2014.items(), key=operator.itemgetter(1), reverse=True)
#print(sorted_dict)

# find the highest ranking terror countries since the past decade
terr_2014 = {}
for each in all_attacks:
    country = each["country_txt"]
    year = float(each["iyear"])
    if year > 2004:
        if country in terr_2014:
            terr_2014[country] += 1
        else:
            terr_2014[country] = 1
            
countries_gdp , countries_terror = [], []
sorted_terror = sorted(terr_2014.items(), key=operator.itemgetter(1), reverse=True)
rank_gdp, rank_terror = {}, {}

for index in range(len(sorted_dict)):
    countries_gdp.append(sorted_dict[index][0])
    rank_gdp[sorted_dict[index][0]] = index+1
    
for index in range(len(sorted_terror)):
    countries_terror.append(sorted_terror[index][0])
    rank_terror[sorted_terror[index][0]] = index+1
    
for each in sorted_terror:
    countries_terror.append(each[0])
    
set_gdp = set(countries_gdp)
set_terr = set(countries_terror)
#print(set_gdp)
#print(set_terr)

# find the common countries and compare their positions
common_countries = (set(countries_gdp)).intersection(set(countries_terror))
#print(common_countries)

print("country", " "*15, "terror rank   ", "GDP rank")
X, Y = [], []
for country in common_countries:
    terr_rank = rank_terror[country]
    gdp_rank  = rank_gdp[country]
    
    if (gdp_rank < 50 or terr_rank < 50):
        print(format(country, "30"), format(terr_rank, "6"), format(gdp_rank, "6"))
        X.append(terr_rank)
        Y.append(gdp_rank)
```

Lets do a simple scatter plot plotting, taking terror rank on X-axis and GDP rank on Y-axis

```python
plt.plot(X,Y, 'co')

slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(X, Y)
line = [ slope*xi + intercept for xi in X]


import scipy.stats
y_bins,bin_edges, misc = scipy.stats.binned_statistic(X,Y,
                                                      statistic="mean", bins=8)
x_bins = (bin_edges[:-1]+bin_edges[1:])/2


plt.hold(True)
plt.plot(x_bins, y_bins, "ro-", linewidth=5)

plt.title("Plot of GDP index and Terrorism Index")
plt.xlabel("Terror Rank")
plt.ylabel("GDP rank")

plt.savefig("int/gdp_and_terrorism_relation.png")
plt.show()
```
